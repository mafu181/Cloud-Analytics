```python
import pandas as pd
df = pd.read_csv("train.csv")
df
```

**Answer of question 1**

```python
trainEnglish = df["English"].median()
df["English"].fillna(trainEnglish, inplace=True)
df
```

```python
trainScience = df["Science"].median()
df["Science"].fillna(trainScience, inplace= True)
df
```

```python
df
```

**After applying one-hot-encoding and imputation with median**

```python
df = pd.get_dummies(df, columns=['School'])
df
```

**Preparing the test Data set from here**

```python
df_test = pd.read_csv('test.csv')
df_test
```

```python
testEnglish = df['English'].median()
df_test['English'].fillna(testEnglish, inplace=True)
df_test
```

```python
testScience = df['Science'].median()
df_test['Science'].fillna(testScience, inplace=True)
df_test
```

# **Answer of the question No 5**

**MSE, MAE, R2 on Training Data Set**

```python
from sklearn.metrics import mean_squared_error
predicted_train_math = reg.predict(df[['English', 'Science']])
actual_train_math = df['Math']
mse_err_train = mean_squared_error(actual_train_math, predicted_train_math)
print('Mean Squared Error on Training Data Set (MSE):',mse_err_train)
```

```python
from sklearn.metrics import mean_absolute_error
mae_err_train = mean_absolute_error(actual_train_math, predicted_train_math)
print('Mean Absolute Error (MAE) on Training Data:', mae_err_train)
```

```python
from sklearn.metrics import r2_score
r2_err_train = r2_score(actual_train_math, predicted_train_math)
print("R2 Score on Training Data Set:", r2_err_train)
```

**MSE, MAE, R2 On Test Data Set**

```python
from sklearn.metrics import mean_squared_error

predicted_math = reg.predict(df_test[['English', 'Science']])
actual_math = df_test['Math']
mse_err = mean_squared_error(actual_math, predicted_math)
print("Mean Squared Error (MSE):",mse_err)
```

```python
from sklearn.metrics import mean_absolute_error

mae_err = mean_absolute_error(actual_math, predicted_math)
print("Mean Absolute Error (MAE):", mae_err)
```

```python
from sklearn.metrics import r2_score

r2_err = r2_score(actual_math, predicted_math)
print("R2 Score:", r2_err)
```
